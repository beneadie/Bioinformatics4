#### Unweighted functions

```python
def most_common(arr):
    return max(set(arr), key=arr.count)

def five_near_unweighted(arr1, arr2):
    new_str=""
    #first element
    temp = arr2[:3]
    new_str+=most_common(temp)
    #second element
    temp = arr2[:4]
    new_str+=most_common(temp)
    #others
    for i in range(2, len(arr1)-2):
        temp = arr2[i-2:i+3]
        new_str+=most_common(temp)
        
    #seond last element
    temp = arr2[-4:]
    new_str+=most_common(temp)
    #last element
    temp = arr2[-3:]
    new_str+=most_common(temp)
    
    return new_str

def three_near_unweighted(arr1, arr2):
    new_str=""
    #first element
    temp = arr2[:2]
    new_str+=most_common(temp)
    #others
    for i in range(1, len(arr1)-1):
        temp = arr2[i-1:i+2]
        new_str+=most_common(temp)
    #last element
    temp = arr2[-2:]
    new_str+=most_common(temp)
    
    return new_str
         
```

```python
import numpy as np

def weighted_choose_5(arr):
    uniqs = list(set(arr))
    scores = (np.zeros_like(uniqs, dtype=int)).tolist() # has to be normal list to rewrite elements
    #print(scores)
    weights = {0: 0.25, 1:0.5, 2:0.65, 3:0.5, 4:0.25}
    for i in range(0, len(uniqs)):
        for j in range(0, len(arr)):
            if uniqs[i] == "X":
                pass
            elif uniqs[i] == arr[j]:
                scores[i] += weights[j]
                #print("match", arr[j], uniqs[i], scores[i])
    #print(scores)
    return uniqs[np.argmax(scores)]


def weighted_choose_3(arr):
    uniqs = list(set(arr))
    scores = (np.zeros_like(uniqs, dtype=int)).tolist()# has to be normal list to rewrite elements
    #print(scores)
    weights = {0: 0.3, 1:0.45, 2:0.3}
    for i in range(0, len(uniqs)):
        for j in range(0, len(arr)):
            if uniqs[i] == "X":
                pass
            elif uniqs[i] == arr[j]:
                scores[i] += weights[j]
                #print("match", arr[j], uniqs[i], scores[i])
    #print(scores)
    return uniqs[np.argmax(scores)]


def weighted_5(arr1, arr2):
    new_str=""
    #first element
    temp = "XX" + arr2[:3]
    new_str+=weighted_choose_5(temp)
    #second element
    temp = "X" + arr2[:4]
    new_str+=weighted_choose_5(temp)
    #others
    for i in range(2, len(arr1)-2):
        temp = arr2[i-2:i+3]
        new_str+=weighted_choose_5(temp)
        
    #seond last element
    temp = arr2[-4:] +"X"
    new_str+=weighted_choose_5(temp)
    #last element
    temp = arr2[-3:] +"XX"
    new_str+=weighted_choose_5(temp)
    
    return new_str

def weighted_3(arr1, arr2):
    new_str=""
    #first element
    temp = "X"+ arr2[:2]
    new_str+=weighted_choose_3(temp)
    #others
    for i in range(1, len(arr1)-1):
        temp = arr2[i-1:i+2]
        new_str+=weighted_choose_3(temp)
    #last element
    temp = arr2[-2:]+ "X"
    new_str+=weighted_choose_3(temp)
    
    return new_str
         
```

```python
aa_combinations_df_base =pd.DataFrame({'Amino Acid Combination': pd.Series(dtype='str'), 'Secondary Structure': pd.Series(dtype='str'),'count': pd.Series(dtype='int')})
aa_combinations_df_base
```

```python
def making_counts(df, outdf):     # variables coming in
    
    #set up data holding 
    combos = [] # originally was dicitoinary to run faster but "not in"   statement did not work perfectly
    counts = {}
                
    for index, row in df.iterrows():
        AA_str = row["AA_seq"]
        SS_str = row["SS"]
        
        # edge case of first and last in string
        special_comb = ("-" +AA_str[:2] + SS_str[0])
        if special_comb not in combos:
            combos.append(special_comb) 
            counts[special_comb] = 1
        else:
            counts[special_comb] = counts.get(special_comb,0) + 1
        special_comb = (AA_str[-2:]+"-"+SS_str[-1])
            
        if special_comb not in combos:
            combos.append(special_comb)
            counts[special_comb] = 1
        else:
            counts[special_comb] = counts.get(special_comb,0) + 1
            
                    
        #normal string combinations
        for i in range(1, len(AA_str)-1):
            key = AA_str[i-1:i+2] + SS_str[i]
            if key not in combos:
                combos.append(key)
                counts[key] = 1
            else:
                counts[key] = counts.get(key,0) + 1
            
    
            
    # making the db
    for co in combos:
        aa = co[:3]
        ss = co[-1]
        
        outdf.loc[len(outdf.index)] = [aa, ss, 0]
    
    for k in counts:
        c = counts[k]
        aa = k[:3]
        ss = k[-1]
        
        outdf["count"].loc[(outdf['Amino Acid Combination'] == aa) & (outdf['Secondary Structure'] == ss)] = c
                
    return outdf
```
