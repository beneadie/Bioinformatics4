```python
import pandas as pd
```

```python
df = pd.read_csv("./Assignment 4 attached files 20 April 2023 312 PM/trainCounts.txt", sep="\t", header=None)
df['AA'] = df[0].astype(str).str[0]
df['SS'] = df[0].astype(str).str[1]
df = df.drop(columns=[0])
df = df.rename({1: "count"}, axis=1)
df
```

```python
aavals = pd.unique(df['AA'])
results = []

def get_perc(m, arr):
    return m/sum(arr)
    

guessmap = dict()



for i in aavals:
    temp_df = df.loc[df["AA"]==i]
    arr1 = temp_df["count"].values.tolist()
    m = max(arr1)
    most_common = temp_df["SS"].loc[temp_df['count'] == m].values
    prob = most_common[0]
    
    guessmap[i] = prob
    
    results.append(get_perc(m, arr1))

    print(f"Amin Acid:  {i},\tSecondary Structure:  {prob} \nProbability: {get_perc(m, arr1)}\n")
     
    
```
